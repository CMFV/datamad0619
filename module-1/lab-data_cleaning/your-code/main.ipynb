#### 11. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
#  quitaria todos los decimales 
clean_merge.dtypes
```

```python
import numpy as np
columns = ['postId', 'Score', 'ViewCount', 'CommentCount']
clean_merge[columns] = clean_merge[columns].applymap(np.int64)

clean_merge.head()
```

```python
stats = clean_merge.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outliers = pd.DataFrame(columns=Merge_df.columns)

for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = clean_merge[(clean_merge[col] < lower) | 
                   (clean_merge[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)

display(outliers)
    
```
