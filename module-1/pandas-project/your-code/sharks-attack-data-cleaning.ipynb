```python
def categorizacion_actividades(m):
    m = str(m)
    if (re.search('\s*([Dd]iving)|([Dd]ive)|([Ss]norkeling)|([Ss]norkel)\s*', m)):
        return 'Dive_sport'
    elif (re.search('\s*([Cc]ollision)|([Aa]ccident)|([Cc]rashed)|([Tt]orpeded)|([Cc]rash)|([Dd]isaster)\s*', m)):
        return 'Sea_accident'
    elif (re.search('\s*([Ff]ishing)|([Cc]atching)|([Ff]ish)|([Cc]ollecting)|([Cc]rabbing)|([Ll]obstering)|([Ss]hrimp)|([Hh]unting) \s*', m)):
        return 'Fishing_activity'   
    elif (re.search('\s*([Bb]oat)|([Ss]hip)|([Oo]nboard)|([Oo]verboard)|([Ss]ailing)|([Rr]iding)|([Cc]ruising)\s*', m)):
        return 'Browsing'
    elif (re.search('\s*([Ss]urfing)|([Bb]oarding)|([Ss]urf)|([Bb]oard)|([Pp]addl)\s*', m)):
        return 'Surf_sport'
    elif (re.search('\s*([Kk]ayaking)|([Kk]ayak)|([Cc]anoeing)|([Rr]owing)\s*', m)):
        return 'Rowing_sports'
    elif (re.search('\s*([Ss]wimming)|([Ss]wim)|([Ss]tanding)|([Ww]alking)|([Ww]ading)|([Tt]reading)|([Bb]each)|([Ff]loating)|([Bb]athing)|([Pp]laying)|([Jj]ump)|([Ss]hore)\s*', m)):
        return 'Leisure_beach'
    elif (re.search('\s*([Ss]hark)|([Bb]rowsing) \s*', m)):
        return 'Shark_activity'
    else:
        return 'Others'


shark_attack_1['Activity']=shark_attack_1['Activity'].apply(categorizacion_actividades)

display(shark_attack_1)
```

```python
def ataques_muerte(x):
    x = str(x)
    if (re.search('(UNKNOWN|Unknown)', x)):
        return 'Unknown'
    elif(re.search('\s*[Nn]|[F]\s*', x)):
        return 'NO'
    elif (re.search('\s*[Y]\s*', x)):
        return 'YES'
    
    
shark_attack_1['Fatal (Y/N)']=shark_attack_1['Fatal (Y/N)'].apply(ataques_muerte)
display(shark_attack_1)
```

```python
def especies_tiburon(s):
    s = str(s)
    if(re.search('\s*([Ww]hitetip)\s*', s)):
        return 'OCEANIC WHITETIP SHARK '
    elif(re.search('\s*([Ww]hite)\s*', s)):
        return 'WHITE SHARK'
    elif(re.search('\s*([Tt]iger)|([Rr]aggedtooth) \s*', s)):
        return 'TIGER SHARK'
    elif (re.search('\s*([Bb]ull)\s*', s)):
        return 'BULL SHARK'
    elif (re.search('\s*([Ss]hortfin)|([Mm]ako)\s*', s)):
        return 'SHORTFIN MAKO SHARK'
    elif (re.search('\s*([Ll]emon)\s*', s)):
        return 'LEMON SHARK'
    elif (re.search('\s*([Bb]lue)\s*', s)):
        return 'BLUE SHARK'
    elif (re.search('\s*([Gg]alapago)\s*', s)):
        return 'GALAPAGOS SHARK'
    elif (re.search('\s*([Cc]aribbean)\s*', s)):
        return 'CARIBBEAN REEF SHARK'
    elif (re.search('\s*([Dd]usky)\s*', s)):
        return 'DUSKY SHARK'
    elif (re.search('\s*([Bb]lack)\s*', s)):
        return 'BLACKTIP SHARK'
    elif (re.search('\s*([Ss]ilk)\s*', s)):
        return 'SILKY SHARK'
    elif (re.search('\s*([Gg]ray)\s*', s)):
        return 'GRAY REEF SHARK'
    elif (re.search('\s*([Hh]ammerhead)\s*', s)):
        return 'GREAT HAMMERHEAD SHARK'
    elif (re.search('\s*([Bb]roadnose)\s*', s)):
        return 'BROADNOSE SEVENGILL SHARK'
    elif (re.search('\s*([Bb]luntnose)|([Ss]ixgill)\s*', s)):
        return 'BLUNTNOSE SIXGILL SHARK'
    elif (re.search('\s*([Nn]urse)\s*', s)):
        return 'NURSE SHARK'
    elif (re.search('\s*([Ss]and)|(DUSKY)|([Dd]usky)\s*', s)):
        return 'SAND TIGER'
    elif (re.search('\s*([Ss]potted)|([Ww]obbegong)\s*', s)):
        return 'SPOTTED WOBBEGONG'
    elif (re.search('\s*([Bb]asking)\s*', s)):
        return 'BASKING SHARK'
    elif (re.search('\s*([Ss]pinner)\s*', s)):
        return 'SPINNER SHARK'
    elif (re.search('\s*([Bb]ronze)|([Cc]opper)\s*', s)):
        return 'BRONZE WHALER SHARK'
    elif (re.search('\s*([Tt]hresher)\s*', s)):
        return 'THRESHER SHARK'
    elif (re.search('\s*([Ww]ale)\s*', s)):
        return 'WHALE SHARK'
    elif (re.search("\s*([21-26] ft)|([21-26]')|([6-8] m)|([Bb]ig)|([Ll]arge)|([Hh]uge)\s*", s)):
        return 'BIG SHARK'
    elif (re.search("\s*([7 - 21] ft)|([7 - 21]')|([2-6] m)|([Mm]edium)\s*", s)):
        return 'MEDIUM SHARK'
    elif (re.search("\s*([0-7] ft)|([0-7]')|([0-2] m)|([Ss]mall)\s*", s)):
        return 'SMALL SHARK'
    else:
        return 'OTHERS' 
shark_attack_1['Species ']=shark_attack_1['Species '].apply(especies_tiburon)
display(shark_attack_1)
```
