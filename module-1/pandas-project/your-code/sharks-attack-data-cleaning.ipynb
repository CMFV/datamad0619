```python
def categorizacion_actividades(m):
    m = str(m)
    if (re.search('\s*([Dd]iving)|([Dd]ive)|([Ss]norkeling)|([Ss]norkel)\s*', m)):
        return 'Dive_sport'
    elif (re.search('\s*([Cc]ollision)|([Aa]ccident)|([Cc]rashed)|([Tt]orpeded)|([Cc]rashed)\s*', m)):
        return 'Sea_accident'
    elif (re.search('\s*([Ff]ishing)|([Cc]atching)|([Ff]ish)|([Cc]olecting)\s*', m)):
        return 'Fishing_activity'   
    elif (re.search('\s*([Bb]oat)|([Ss]hip)|([Oo]nboard)|([Oo]verboard)\s*', m)):
        return 'Browsing'
    elif (re.search('\s*([Ss]urfing)|([Bb]oarding)|([Ss]urf)|([Bb]oard)\s*', m)):
        return 'Surf_sport'
    elif (re.search('\s*([Kk]ayaking)|([Kk]ayak)|([Cc]anoeing)|([Rr]owing)\s*', m)):
        return 'Rowing_sports'
    elif (re.search('\s*([Ss]wimming)|([Ss]wim)|([Ss]tanding)|([Ww]alking)|([Ww]ading)|([Tt]reading)|([Bb]each)|([Ff]loating)\s*', m)):
        return 'Leisure_beach'
    else:
        return m


shark_attack_1['Activity']=shark_attack_1['Activity'].apply(categorizacion_actividades)

shark_attack_1.head(100)
```
